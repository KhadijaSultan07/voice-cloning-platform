```python
# ============================================
# GRADIO APP - PUBLIC URL
# ============================================

!pip install gradio

import gradio as gr
from voxcpm import VoxCPM
import soundfile as sf
import tempfile
import os

print("🔄 Loading model...")
model = VoxCPM.from_pretrained("./voxcpm2_model", load_denoiser=False)
print("✅ Model loaded!")

def generate_voice(audio_file, text):
    try:
        if audio_file is not None:
            # Convert uploaded file to WAV if needed
            if audio_file.endswith('.ogg'):
                from pydub import AudioSegment
                audio = AudioSegment.from_ogg(audio_file)
                wav_path = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name
                audio.export(wav_path, format="wav")
                ref_audio = wav_path
            else:
                ref_audio = audio_file

            wav = model.generate(text, reference_wav_path=ref_audio, cfg_value=3.0, inference_timesteps=25)
        else:
            wav = model.generate(text, cfg_value=3.0, inference_timesteps=25)

        temp_path = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name
        sf.write(temp_path, wav, 48000)
        return temp_path
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio Interface
with gr.Blocks(title="Cross-Lingual Voice Cloning", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # 🎙️ Cross-Lingual Voice Cloning Studio
    ### Upload any voice → Generate in any language (Urdu, English, Arabic, etc.)
    """)

    with gr.Row():
        with gr.Column():
            audio_input = gr.Audio(label="🎤 Upload Voice (Optional)", type="filepath")
            text_input = gr.Textbox(
                label="📝 Text (Any Language)",
                placeholder="Assalam-o-Alaikum! Yeh mera test hai.",
                lines=3
            )
            generate_btn = gr.Button("🎵 Generate Voice", variant="primary", size="lg")

        with gr.Column():
            audio_output = gr.Audio(label="🔊 Generated Audio", type="filepath")
            gr.Markdown("""
            ### 💡 Tips:
            - **Voice Cloning:** Upload voice + Text
            - **Voice Design:** Text with description *(e.g., "(A female voice) Hello!")*
            - **Cross-Lingual:** Any language input works
            """)

    generate_btn.click(generate_voice, [audio_input, text_input], audio_output)

# Public link
demo.launch(share=True, server_name="0.0.0.0")
```

```python
# ============================================
# COMPLETE GRADIO APP - PUBLIC URL
# ============================================

import gradio as gr
from voxcpm import VoxCPM
import soundfile as sf
import tempfile

print("🔄 Loading model...")
model = VoxCPM.from_pretrained("./voxcpm2_model", load_denoiser=False)
print("✅ Model loaded!")

def generate_voice(text, audio_file):
    try:
        if audio_file is not None:
            wav = model.generate(
                text=text,
                reference_wav_path=audio_file,
                cfg_value=2.0,
                inference_timesteps=15,
                denoise=False
            )
        else:
            wav = model.generate(
                text=text,
                cfg_value=2.0,
                inference_timesteps=15,
                denoise=False
            )
        temp_path = tempfile.NamedTemporaryFile(delete=False, suffix=".wav").name
        sf.write(temp_path, wav, 48000)
        return temp_path
    except Exception as e:
        return f"Error: {str(e)}"

# Gradio Interface
with gr.Blocks(title="Urdu Voice Cloning", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # 🎙️ Urdu Voice Cloning Studio
    ### 30+ Languages | 48kHz Quality | Free to Use
    """)

    with gr.Row():
        with gr.Column(scale=1):
            text_input = gr.Textbox(
                label="📝 Text",
                placeholder="Urdu, English, Arabic... kisi bhi language mein text likhein...",
                lines=4
            )
            audio_input = gr.Audio(
                label="🎤 Reference Audio (Optional - Voice Cloning)",
                type="filepath"
            )
            generate_btn = gr.Button("🎵 Generate Voice", variant="primary", size="lg")

        with gr.Column(scale=1):
            audio_output = gr.Audio(label="🔊 Generated Audio", type="filepath")
            gr.Markdown("""
            ### 💡 Tips:
            - **Simple TTS:** Sirf text likhein aur Generate karein
            - **Voice Cloning:** Reference audio upload karein + text
            - **Voice Design:** Parentheses mein description likhein
              *(e.g., "(A calm, natural voice) Assalam-o-Alaikum!")*
            """)

    generate_btn.click(generate_voice, [text_input, audio_input], audio_output)

# Public link
demo.launch(share=True, server_name="0.0.0.0")
```
